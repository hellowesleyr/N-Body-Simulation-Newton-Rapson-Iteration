PV = NKT is a well known equation which explains the behaviour of an ideal gas.
Given that interactions for particles within proximtiy of 1 sigma are disabled, I suppose this should apply to the argon gas. One property of the ideal gas is that the particles are infinitesimal in volume, this condition is aproximated by the previous rule, as it suggests that particles cannot collide.

```python
#Drawing from the data set where bos length was varied;
fig,ax = plt.subplots()
ax.set_title('Pressure volume product gainst temperature')
ax.set_ylabel('Pressure * volume (J)')
ax.set_xlabel('N * Average Temperature (J)')
ax.plot(box_volumes[5:]*inv1_pressures[5:],inv1_temperatures[5:])
```

Were aiming to see linear behaviour throughout the whole graph. We see it briefly before we elevate the box length to 50*sigma at which point the relationship appears to take an exponential increase.
This can be explained by an insufficient length of runtime for the simulation. As the box length increases, it takes longer for particles to begin interacting with the box walls. If the simulation cut off soon after the first wall collisions, this would yield an artificially elevated pressure*volume product. This is the strongest explanation for the observed phenomena, as during the data gathering phase, I had to increase the simulation runtime as originally negligible values for pressure were being recorded for large volume situations. Trivially this was remedied by increasing the simulation length, but it appears that the side effect of artifically raising pressure\*volume poroduct were not visible until now.

Accordingly, if we take the above hypothese for unphysical data to be true, we obtain a new graph:


```python
#Drawing from the data set where bos length was varied;
PV_Arr = box_volumes*inv1_pressures
fig,ax = plt.subplots()
ax.set_title('Pressure volume product gainst temperature')
ax.set_ylabel('Pressure * volume (J)')
ax.set_xlabel('N * Average Temperature (J)')
ax.plot(PV_Arr[7:12],inv1_temperatures[7:12])
```

We would expect the gradient of this section to equal N*R, where N,100, is the particle number, and R is the molar gas constant

```python
gradient = np.gradient(PV_Arr[7:12],inv1_temperatures[7:12])
np.mean(gradient)
```
